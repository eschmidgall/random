```python
#Flow, mean is 1.53, stdev is 0.3
#Linewidth, mean is 13.65, stdev is 0.08
#Pressure,mean is 1500, stdev is 100
#Temperature, mean is 100, stdev is 10
mnList = [1.53, 13.65, 1500, 100]
stList = [0.3, 0.08, 100, 10]
quantList = ['flow', 'linewidth', 'pressure', 'temperature']

for q in range(4):
    mn = mnList[q]
    std = stList[q]
    quant =  quantList[q]
    for n in range(10):
        timeVec = []
        for m in range(100):
            now = datetime.now()
            dt_string = now.strftime("%d/%m/%Y %H:%M:%S.%f")
            timeVec.append(dt_string)
            time.sleep(0.02)
        flow = np.random.normal(mn, std, 100)
        rule = np.random.choice([0,1,2,3,4])
        mval = np.mean(flow)
        stdval = np.std(flow)

        print(f'File {quant}_{n}.csv will have rule {rule} applied. Actual mean {mval} vs target {mn}, actual stdev {stdval} vs target {std}')

        if rule ==0: 
            elemnt = 0
            val = 0

        if rule == 1: 
            elemnt = np.random.randint(0,100)
            sgn = np.random.choice([-1,1])
            if sgn>0: 
                val = np.random.uniform(mval+3*stdval, mval+5*stdval)
            if sgn<0: 
                val =np.random.uniform(mval-3*stdval, mval-5*stdval)
            flow[elemnt]= val
        
        if rule == 2: 
            elemnt = np.random.randint(0,99)
            sgn = np.random.choice([-1,1])
            if sgn>0: 
                val = np.random.uniform(mval+2*stdval, mval+3*stdval, 2)
            if sgn<0: 
                val = np.random.uniform(mval-3*stdval, mval-2*stdval, 2)
            flow[elemnt]= val[0]
            flow[elemnt+1]=val[1]
        
        if rule == 3: 
            elemnt = np.random.randint(0,95)
            sgn = np.random.choice([-1,1])
            if sgn>0: 
                val = np.random.uniform(mval+1*stdval, mval+2*stdval, 4)
            if sgn<0: 
                val = np.random.uniform(mval-2*stdval, mval-1*stdval, 4)
            flow[elemnt]= val[0]
            flow[elemnt+1]=val[1] 
            flow[elemnt+2]=val[2]
            flow[elemnt+3]=val[3]
        
        if rule ==4: 
            num = np.random.randint(9, 15)
            elemnt = np.random.randint(0,100-num)
            sgn = np.random.choice([-1,1])
            if sgn>0: 
                val = np.random.uniform(mval, mval+1*stdval, num)
            if sgn<0: 
                val = np.random.uniform(mval-1*stdval, mval, num)
            for mm in range(num): 
                flow[elemnt+mm]= val[mm]


        data = [[a,b] for a,b in zip(timeVec, flow)]
        with open(f'{quant}_1{n}.csv', 'w', newline='') as file: 
            writer = csv.writer(file)
            writer.writerows(data)
        
        master_list.append([quant, n, rule, elemnt, val ])



```

```python
file_list = glob.glob('*.csv')
for file in file_list: 
    if 'master_list' in file: 
        continue
    else: 
        print(f'Now processing {file}.') 
        tseries = []
        data = []
        with open(file) as csv_file: 
            csv_reader = csv.reader(csv_file)
            for row in csv_reader: 
                tseries.append(row[0])
                data.append(float(row[1]))
        #Rule evaluation
        rules = [False, False, False, False, False]
        mVal = np.mean(data)
        sdVal = np.std(data)
        #print(f'Mean in file {file} is {mVal}, stdev {sdVal}')
        #print(f'Rule 3 control limits are {mVal+sdVal} and {mVal-sdVal}')
        #print(f'Max is {np.max(data)} and min is {np.min(data)}')

        n=0 
        for item in data: 
            if item > mVal+3*sdVal or item <mVal-3*sdVal: 
                if rules[1]==False:
                    rules[1]=True
                    print(f'File {file} satisfies rule 1.')
            if item>mVal+2*sdVal and data[n+1]>mVal+2*sdVal: 
                if rules[2]==False: 
                    rules[2] = True
                    print(f'File {file} satisifies rule 2. ')
            if item<mVal-2*sdVal and data[n+1]<mVal-2*sdVal: 
                if rules[2] == False: 
                    rules[2] = True
                    print(f'File {file} satisfies rule 2.')
            if item>mVal+sdVal: 
                counter = 1
                sequential = True
                while sequential: 
                    if n+counter+1 >len(data):
                        sequential = False
                    elif data[n+counter]>mVal+sdVal: 
                        counter+=1
                    else: 
                        sequential = False
                if counter>=4: 
                    print(f'There are {counter} data points in a row above 1 sigma')
                    if rules[3]==False: 
                        rules[3] = True
                        print(f'File {file} satisfies rule 3.')
            if item<mVal-sdVal: 
                counter = 1
                sequential = True
                while sequential: 
                    if n+counter+1 >len(data):
                        sequential = False
                    elif data[n+counter]<mVal-sdVal: 
                        counter+=1
                    else: 
                        sequential = False
                if counter>=4: 
                    print(f'There are {counter} data points in a row below 1 sigma')
                    if rules[3]==False: 
                        rules[3] = True
                        print(f'File {file} satisfies rule 3.')
            if item>mVal: 
                    counter = 1
                    sequential = True
                    while sequential: 
                        if n+counter+1 >len(data):
                            sequential = False
                        elif data[n+counter]>mVal: 
                            counter+=1
                        else: 
                            sequential = False
                    if counter>=9: 
                        if rules[4]==False: 
                            rules[4] = True
                            print(f'There are {counter} data points in a row above the mean')
                            print(f'File {file} satisfies rule 4.')
            if item<mVal: 
                    counter = 1
                    sequential = True
                    while sequential: 
                        if n+counter+1 >len(data):
                            sequential = False
                        elif data[n+counter]<mVal: 
                            counter+=1
                        else: 
                            sequential = False
                    if counter>=9: 
                        if rules[4]==False: 
                            rules[4] = True
                            print(f'There are {counter} data points in a row below the mean')
                            print(f'File {file} satisfies rule 4.')
            n+=1


```
